### Check the distribution of packet loss values for each of the source sites.

```python
rows = int(np.ceil(len(sdatalist)/2))
f, axes = plt.subplots(rows, 2, figsize=(20, 27))
i = 0
for r, c in list(itertools.product(range(0,rows), range(0,2))):
    if i < len(sdatalist):
        f = sns.histplot(ax=axes[r, c], data=sdatalist[i], x="packet_loss",bins=10,stat='probability').set_title(sdatalist[i]['src_site'].unique(), y=0.1, fontsize = 12)
    i += 1
```

#### Check the worst 20 destinations for each selected source

```python
f, axes = plt.subplots(rows, 2, figsize=(25, 42))

i=0
for r, c in list(itertools.product(range(0,rows), range(0,2))):
    if i < len(sdatalist):
#         print(sdatalist[i]['src_site'].unique())
        withloss = sdatalist[i][sdatalist[i]['packet_loss']>0.5]['src'].unique()
        extracted = sdatalist[i][sdatalist[i]['src'].isin(withloss)]
        grouped = extracted.groupby(['dest_host', 'dest_site']).mean().round(decimals=3).reset_index()
        grouped = grouped.sort_values('packet_loss', ascending=False).head(20)
        sns.barplot(ax=axes[r, c], y="dest_site", x="packet_loss", data=grouped).set_title(sdatalist[i]['src_site'].unique())
    i+=1
```

```python
f.savefig("selected_sources_worst20_destinations.png")
```
